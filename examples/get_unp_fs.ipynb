# Example: Getting Financial Statements For UNP

- Company name: Union Pacific Corporation (UNP) 
- Fiscal Year: 2021
- Filing Year: 2022


## The goal is to get the financial statements for a given company and year

It takes some effort to find the URL of the financial statements for this particular company and year. 

<div style="color: blue;">But how about getting the financial statements for <b>any company and year</b>, and do it quickly?</div>



```python
import requests
from IPython.display import HTML

# Fetch the content
response = requests.get('https://www.sec.gov/Archives/edgar/data/100885/000143774922002494/R5.htm', 
                    headers={"user-agent": "Your Name youremail@your.domain.com"})
html_content = response.text

# Display the HTML content
HTML(html_content)
```

```python
import pandas as pd
from sec_xbrl.get_statement import  get_annual_statement
from sec_xbrl.filing import Filing

# Customize cache settings
from pathlib import Path
Filing.CACHE_DIR = Path('sec_reports')  # Change cache directory
Filing.CACHE_VALIDITY_DAYS = 7  # Change cache expiration to 7 days
```

```python
df_all = get_annual_statement(ticker="UNP", year=2022)
```

```python
df_balance = df_all[df_all["statement"]=="BalanceSheet"]
df_balance.to_csv("UNP_BalanceSheet_2022.csv", index=False)
df_balance[["xbrl_tag",	"label",	"value_in_millions"]].head(60)
```

```python
df_spf = df_all[df_all["statement"]=="IncomeStatement"]
df_spf.to_csv("UNP_IncomeStatement_2022.csv", index=False)
df_spf[["xbrl_tag",	"label",	"value_in_millions"]].head(60)
```
